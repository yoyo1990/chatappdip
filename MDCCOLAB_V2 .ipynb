```python
# ייבוא ספריות נדרשות כבר בתחילת הסקריפט
import os # חובה לייבא את os לפני השימוש בו
import zipfile
from google.colab import files
import shutil # לייבוא ספריית shutil עבור העברת תיקיות

# התקנת ספריית Kaggle
!pip install kaggle

# יצירת תיקייה ל-API key של Kaggle
!mkdir -p ~/.kaggle

# העלאת קובץ kaggle.json
# Colab יבקש ממך לבחור את הקובץ מהמחשב שלך. בחר את קובץ ה-kaggle.json שהורדת.
print("אנא העלה את קובץ ה-kaggle.json שלך כעת:")
uploaded = files.upload()

# העברת קובץ ה-kaggle.json לתיקייה הנכונה
!mv kaggle.json ~/.kaggle/kaggle.json
!chmod 600 ~/.kaggle/kaggle.json

# הגדרות תחרות
COMPETITION_NAME = 'make-data-count-finding-data-references'

# נתיבי יעד סופיים (מבנה Kaggle)
KAGGLE_INPUT_DIR = f'/content/kaggle/input/{COMPETITION_NAME}'
MAIN_COMPETITION_ZIP_PATH = f'{COMPETITION_NAME}.zip'

# --- שלבי ניקוי לפני הורדה ופריסה ---
print("מנקה תיקיות וקבצים קודמים כדי למנוע כפילויות וקינון יתר...")
if os.path.exists(KAGGLE_INPUT_DIR):
    shutil.rmtree(KAGGLE_INPUT_DIR)
    print(f"נמחק תיקיית יעד קודמת: {KAGGLE_INPUT_DIR}")
if os.path.exists(MAIN_COMPETITION_ZIP_PATH):
    os.remove(MAIN_COMPETITION_ZIP_PATH)
    print(f"נמחק קובץ ZIP ראשי קודם: {MAIN_COMPETITION_ZIP_PATH}")
# ניקוי תיקיית temp ישנה אם קיימת מהרצות קודמות עם לוגיקה שונה
legacy_temp_dir = f'/content/{COMPETITION_NAME}_temp'
if os.path.exists(legacy_temp_dir):
    shutil.rmtree(legacy_temp_dir)
    print(f"נמחק תיקיית temp ישנה: {legacy_temp_dir}")
print("ניקוי הושלם.")
# --- סוף שלבי ניקוי ---


# יצירת תיקייה לנתוני הקלט במבנה ש-Kaggle מצפה
# חשוב ליצור אותה רק לאחר הניקוי, ובכל מקרה הקוד יצור אותה
# לפני הפריסה ב-extractall
os.makedirs(KAGGLE_INPUT_DIR, exist_ok=True)


print(f"מוריד נתונים עבור תחרות: {COMPETITION_NAME}...")
# הורדת קובץ ה-ZIP הראשי לתיקיית העבודה הנוכחית
!kaggle competitions download -c {COMPETITION_NAME} -p .

print(f"פורס את קובץ ה-ZIP הראשי: {MAIN_COMPETITION_ZIP_PATH} ישירות לתיקיית היעד: {KAGGLE_INPUT_DIR}...")
if os.path.exists(MAIN_COMPETITION_ZIP_PATH):
    with zipfile.ZipFile(MAIN_COMPETITION_ZIP_PATH, 'r') as zip_ref:
        # חילוץ ישירות ל-KAGGLE_INPUT_DIR
        # זה יצור את train/, test/, train_labels.csv וכו' בתוך KAGGLE_INPUT_DIR
        zip_ref.extractall(KAGGLE_INPUT_DIR)
    print("פריסת קובץ ה-ZIP הראשי הושלמה בהצלחה לתיקיית היעד.")

    # מחיקת קובץ ה-ZIP הראשי לאחר החילוץ
    os.remove(MAIN_COMPETITION_ZIP_PATH)
    print(f"נמחק קובץ ה-ZIP הראשי: {MAIN_COMPETITION_ZIP_PATH}")

else:
    print(f"שגיאה: קובץ ה-ZIP הראשי '{MAIN_COMPETITION_ZIP_PATH}' לא נמצא. ודא שההורדה הצליחה.")


print("\nבדיקת מבנה הקבצים לאחר הפריסה ל-KAGGLE_INPUT_DIR:")
# הצגת תוכן תיקיית היעד כדי לוודא שהמבנה נכון
!ls -R {KAGGLE_INPUT_DIR}

# הגדרת נתיבים לבדיקות מפורטות (אותיות רישיות, כפי שנמצאו בפלט הקודם)
TRAIN_DATA_DEST_DIR = os.path.join(KAGGLE_INPUT_DIR, 'train')
TEST_DATA_DEST_DIR = os.path.join(KAGGLE_INPUT_DIR, 'test')

# בדיקה מפורשת של תוכן תיקיות ה-XML וה-PDF בנתיבי היעד
print(f"\nבדיקת תוכן תיקיית XML באימון: {TRAIN_DATA_DEST_DIR}/XML")
!ls -l {TRAIN_DATA_DEST_DIR}/XML

print(f"\nבדיקת תוכן תיקיית PDF באימון: {TRAIN_DATA_DEST_DIR}/PDF")
!ls -l {TRAIN_DATA_DEST_DIR}/PDF

print(f"\nבדיקת תוכן תיקיית XML בבדיקה: {TEST_DATA_DEST_DIR}/XML")
!ls -l {TEST_DATA_DEST_DIR}/XML

print(f"\nבדיקת תוכן תיקיית PDF בבדיקה: {TEST_DATA_DEST_DIR}/PDF")
!ls -l {TEST_DATA_DEST_DIR}/PDF


print("\nהגדרת סביבת Colab והורדת הנתונים הושלמה בהצלחה!")
```
