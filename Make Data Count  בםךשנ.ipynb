```python
# ייבוא ספריות נדרשות כבר בתחילת הסקריפט
import os # חובה לייבא את os לפני השימוש בו
import zipfile
from google.colab import files

# התקנת ספריית Kaggle
!pip install kaggle

# יצירת תיקייה ל-API key של Kaggle
!mkdir -p ~/.kaggle

# העלאת קובץ kaggle.json
# Colab יבקש ממך לבחור את הקובץ מהמחשב שלך. בחר את קובץ ה-kaggle.json שהורדת.
uploaded = files.upload()

# העברת קובץ ה-kaggle.json לתיקייה הנכונה
!mv kaggle.json ~/.kaggle/kaggle.json
!chmod 600 ~/.kaggle/kaggle.json

# הורדת נתוני התחרות מ-Kaggle
# ודא שזהו השם המדויק של התחרות ב-Kaggle
COMPETITION_NAME = 'make-data-count-finding-data-references' # השם עודכן לתחרות הנכונה
print(f"מוריד נתונים עבור תחרות: {COMPETITION_NAME}...")
# הפעלת הורדה עם אופציה --force כדי לוודא הורדה גם אם קבצים קיימים
# והפניית הפלט לתיקיית העבודה הנוכחית של Colab
!kaggle competitions download -c {COMPETITION_NAME} -p .

# נתיב לקובץ ה-ZIP הראשי של התחרות, אשר הורד לתיקיית העבודה הנוכחית
MAIN_COMPETITION_ZIP_PATH = f'{COMPETITION_NAME}.zip'

# יצירת תיקייה זמנית לחילוץ קובץ ה-ZIP הראשי
TEMP_UNZIP_DIR = f'/content/{COMPETITION_NAME}_temp'
os.makedirs(TEMP_UNZIP_DIR, exist_ok=True)

print(f"פורס את קובץ ה-ZIP הראשי של התחרות: {MAIN_COMPETITION_ZIP_PATH} לתיקייה זמנית: {TEMP_UNZIP_DIR}...")
if os.path.exists(MAIN_COMPETITION_ZIP_PATH):
    with zipfile.ZipFile(MAIN_COMPETITION_ZIP_PATH, 'r') as zip_ref:
        zip_ref.extractall(TEMP_UNZIP_DIR)
    print("פריסת קובץ ה-ZIP הראשי הושלמה.")
    # הצגת תוכן התיקייה הזמנית כדי לוודא שקבצי ה-ZIP הפנימיים קיימים
    print(f"תוכן התיקייה הזמנית {TEMP_UNZIP_DIR}:")
    !ls {TEMP_UNZIP_DIR}
else:
    print(f"שגיאה: קובץ ה-ZIP הראשי '{MAIN_COMPETITION_ZIP_PATH}' לא נמצא. ודא שההורדה הצליחה.")


# יצירת תיקייה לנתוני הקלט במבנה ש-Kaggle מצפה
# Kaggle מצפה שקבצי הקלט יהיו בנתיב /kaggle/input/שם_התחרות
# אז אנחנו ניצור מבנה דומה ב-Colab
KAGGLE_INPUT_DIR = f'/content/kaggle/input/{COMPETITION_NAME}'
os.makedirs(KAGGLE_INPUT_DIR, exist_ok=True)

# נתיבים לקבצים המכווצים בתוך התיקייה הזמנית לאחר פריסת ה-ZIP הראשי
TRAIN_ZIP_PATH = os.path.join(TEMP_UNZIP_DIR, 'train.zip')
TEST_ZIP_PATH = os.path.join(TEMP_UNZIP_DIR, 'test.zip')
TRAIN_LABELS_FILE = os.path.join(TEMP_UNZIP_DIR, 'train_labels.csv')

# יצירת נתיבי יעד עבור הנתונים בתוך KAGGLE_INPUT_DIR
TRAIN_DATA_DIR = os.path.join(KAGGLE_INPUT_DIR, 'train')
TEST_DATA_DIR = os.path.join(KAGGLE_INPUT_DIR, 'test')
TRAIN_LABELS_DEST_PATH = os.path.join(KAGGLE_INPUT_DIR, 'train_labels.csv')

os.makedirs(TRAIN_DATA_DIR, exist_ok=True)
os.makedirs(TEST_DATA_DIR, exist_ok=True)


print("פורס את קבצי האימון...")
# פריסת קובץ train.zip
if os.path.exists(TRAIN_ZIP_PATH):
    with zipfile.ZipFile(TRAIN_ZIP_PATH, 'r') as zip_ref:
        zip_ref.extractall(TRAIN_DATA_DIR) # חולץ לתיקיית train בתוך /kaggle/input/make-data-count/
    print(f"פורס {TRAIN_ZIP_PATH} ל- {TRAIN_DATA_DIR}")
else:
    print(f"אזהרה: קובץ {TRAIN_ZIP_PATH} לא נמצא. ודא שההורדה הצליחה ושהקובץ הראשי נפרס.")


print("פורס את קבצי הבדיקה...")
# פריסת קובץ test.zip
if os.path.exists(TEST_ZIP_PATH):
    with zipfile.ZipFile(TEST_ZIP_PATH, 'r') as zip_ref:
        zip_ref.extractall(TEST_DATA_DIR) # חולץ לתיקיית test בתוך /kaggle/input/make-data-count/
    print(f"פורס {TEST_ZIP_PATH} ל- {TEST_DATA_DIR}")
else:
    print(f"אזהרה: קובץ {TEST_ZIP_PATH} לא נמצא. ודא שההורדה הצליחה ושהקובץ הראשי נפרס.")


# העברת train_labels.csv למקום הנכון
if os.path.exists(TRAIN_LABELS_FILE):
    # יש להעתיק במקום להעביר, אם תרצה לשמור את המקור בתיקייה הזמנית
    !cp {TRAIN_LABELS_FILE} {TRAIN_LABELS_DEST_PATH}
    print(f"העביר {TRAIN_LABELS_FILE} ל- {TRAIN_LABELS_DEST_PATH}")
else:
    print(f"אזהרה: קובץ {TRAIN_LABELS_FILE} לא נמצא, ודא שהורד כראוי ושהקובץ הראשי נפרס.")


print("\nבדיקת מבנה הקבצים לאחר הפריסה:")
!ls -R {KAGGLE_INPUT_DIR}

print("\nהגדרת סביבת Colab והורדת הנתונים הושלמה בהצלחה!")

```
